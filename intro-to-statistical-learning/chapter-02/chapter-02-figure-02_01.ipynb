# FIGURE 2.1 Scatterplot of Advertising Data Scatterplot

```python
# import section 
import matplotlib  # .pyplot, later pyplot might be directly imported
import seaborn
import numpy
import pandas 
```

```python
advertisingDataFrame: pandas.DataFrame = pandas.read_csv ("../islr-data/Advertising.csv")
```

```python
advertisingDataFrame.head (5)
```

```python
advertisingDataFrame.describe()
```

```python

# explicitly declare the type annotations for reference because of the tuple return (out of band declarations)
figure: matplotlib.figure.Figure
figureAxes: numpy.ndarray

# setup a 1 row, 3 column plot layout with same Y scale
figure, figureAxes = matplotlib.pyplot.subplots (1, 3, sharey=True)

# plot the three independent variables against the dependent variable 
# dependent (response) variable is "sales"
# independent (predictor) variables, X_1 is "TV", X_2 is "radio", X_3 is "newspaper"
seaborn.regplot (x = "TV", y = "sales", data = advertisingDataFrame, ax = figureAxes [0], scatter_kws = { "color": "#ff0000" })
seaborn.regplot (x = "radio", y = "sales", data = advertisingDataFrame, ax = figureAxes [1], scatter_kws = { "color": "#ff0000" })
seaborn.regplot (x = "newspaper", y = "sales", data = advertisingDataFrame, ax = figureAxes [2], scatter_kws = { "color": "#ff0000" })
```
